```python
### Section A
# 1. Create a list of 5 integers
nums = [10, 20, 30, 40, 50]
print("Original list:", nums)

# append
nums.append(60)
print("After append:", nums)

# extend
nums.extend([70, 80])
print("After extend:", nums)

# insert
nums.insert(2, 25)
print("After insert:", nums)

# remove
nums.remove(40)
print("After remove:", nums)

# pop
nums.pop()
print("After pop:", nums)

# index
print("Index of 30:", nums.index(30))

# count
print("Count of 20:", nums.count(20))

# sort
nums.sort()
print("After sort:", nums)

# reverse
nums.reverse()
print("After reverse:", nums)

# clear
nums.clear()
print("After clear:", nums)
```

```python
### 7
num = int(input("Enter a number: "))

if num > 0:
    print(f"{num} is positive")
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
elif num == 0:
    print(f"{num} is zero")
else:
    print(f"{num} is negative")
```

```python
## Section C
## 11. Exploring pandas
import pandas as pd

# Create DataFrame manually with 10 rows
data = {
    "Name": ["Aisha", "Yasmeen", "Rahul", "Fatima", "Ravi",
             "Neha", "Aisha", "Imran", "Sara", "Rahul"],
    "Age": [22, 21, 23, 20, 24, 22, 22, 25, 21, 23],
    "City": ["Hyderabad", "Delhi", "Mumbai", "Chennai", "Pune",
             "Delhi", "Hyderabad", "Bangalore", "Mumbai", "Mumbai"],
    "Salary": [45000, 50000, 48000, None, 52000,
               50000, 45000, None, 47000, 48000]
}

df = pd.DataFrame(data)

print("----- Original DataFrame -----")
print(df)

# DataFrame info
print("\n----- DataFrame Info -----")
df.info()

# Statistical description
print("\n----- DataFrame Describe -----")
print(df.describe())

# Select Name and City columns
print("\n----- Selected Columns (Name & City) -----")
print(df[['Name', 'City']])

# Drop City column
df = df.drop(columns=['City'])
print("\n----- After Dropping City Column -----")
print(df)

# Fill missing Salary values with mean
salary_mean = df['Salary'].mean()
df['Salary'].fillna(salary_mean, inplace=True)
print("\n----- After Filling Missing Salary Values -----")
print(df)

# Remove duplicate rows
df = df.drop_duplicates()
print("\n----- After Removing Duplicate Rows -----")
print(df)
```

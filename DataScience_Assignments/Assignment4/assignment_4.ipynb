```python
import pandas as pd
patient_df = pd.read_csv("Patient_Data.csv")
billing_df = pd.read_csv("Billing_Data.csv")
patient_df.info()
billing_df.info()
patient_df['BillAmount'] = patient_df['BillAmount'].fillna(0)
billing_summary = billing_df.groupby(
    'PatientID', as_index=False
)['FinalAmount'].sum()
patient_df = patient_df.drop_duplicates(subset='PatientID')
final_cleaned_df = pd.merge(
    patient_df,
    billing_summary,
    on='PatientID',
    how='left'
)
final_cleaned_df['FinalAmount'] = final_cleaned_df['FinalAmount'].fillna(0)
final_cleaned_df.info()
print(final_cleaned_df)
```

```python
#### Checking for  prime number 
num = int(input("enter the number:"))
if num <= 1:
    print(num, "is not a prime number")
else:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, "is a prime number")
    else:
        print(num, "is not a prime number")
```

```python
####Squares of even numbers
print("Squares of even numbers between 100 and 200:")

for num in range(100, 201):
    if num % 2 == 0:        # Check if the number is even
        print(f"{num}^2 = {num ** 2}")
```

```python
### random numbers
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
answer = int(input(f"What is the product of {num1} and {num2}? "))
if answer == num1 * num2:
    print(" Correct answer!.")
else:
    print("Incorrect.")
    print(f"The correct answer is {num1 * num2}.")
```

```python
#### Hospital billing system
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_charge = 300

days_stayed = int(input("Enter the number of days stayed: "))
number_of_tests = int(input("Enter the number of lab tests: "))
medicine_charges = int(input("Enter the medicine_charges"))

room_charges = days_stayed * room_charge_per_day
lab_charges = number_of_tests * lab_test_charge
```

```python

total_bill = room_charges + doctor_fee + lab_charges + medicine_charges

if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0

final_bill = total_bill - discount

print("Total bill before discount:", total_bill)
print("Discount applied:", discount)
print("Final bill to be paid:", final_bill)
```
